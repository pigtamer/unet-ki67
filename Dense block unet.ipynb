```python
import os
os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"

import keras
from model import *
import segmentation_models as sm


# define model

def denseunet():
    model = sm.Unet('densenet121', encoder_weights=None)
    model.compile(
        'Adam',
        loss=sm.losses.bce_jaccard_loss,
        metrics=[sm.metrics.iou_score],
    )
    return model
```

```python
model = denseunet()
unet = unet()
```
